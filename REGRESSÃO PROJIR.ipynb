```python
x = base.iloc[:, 7].values #DS
y = base.iloc[:, 1].values # ADT
correlacao = np.corrcoef(x, y)
correlacao
# NEGATIVO FORTE
```

```python
# Mudança do formato de X para o formato de matriz (necessário para versões mais recentes do sklearn)
x = x.reshape(-1, 1)
```

```python
# Criação do modelo, treinamento, visualização dos coeficientes e do score do modelo
modelo = LinearRegression()
modelo.fit(x, y)
```

```python
#Interceptação
modelo.intercept_
```

```python
#inclinação
modelo.coef_
```

```python
#score R^2
modelo.score(x, y)
#51% da variável dependente pode ser explicada pelas variávei explanatórias do modelo 
```

```python
# Geração das previsões
previsoes = modelo.predict(x)
previsoes
```

```python
# Criação do modelo, utilizando a biblioteca statsmodel 
#podemos ver r ajustadodo r2
modelo_ajustado = sm.ols(formula = 'ADT ~ DS', data = base)
modelo_treinado = modelo_ajustado.fit()
modelo_treinado.summary()
```

```python
# Visualização dos resultados
plt.scatter(x, y)
plt.plot(x, previsoes, color = 'red')
plt.title( "ADTXDS")
```

```python
from yellowbrick.regressor import ResidualsPlot 

# Gráfico para visualizar os residuais
visualizador = ResidualsPlot(modelo)
visualizador.fit(x, y)
visualizador.poof()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

residuos = y - modelo.predict(x)


plt.figure(figsize=(8, 6))
sns.histplot(residuos, kde=True, color='blue', bins=30)  
plt.xlabel('Resíduos')
plt.ylabel('Contagem')
plt.title('Histograma dos Resíduos com Curva de Densidade')
plt.show()


# resíduos okay
```
